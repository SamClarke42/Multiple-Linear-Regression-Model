```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def B1(x,y):   # Function to compute B1 coefficient.
    
    n = len(x)
    x_bar = np.mean(x)
    y_bar = np.mean(y)
    x_squared_bar = np.mean(x)**2
    total = 0
    x_squared = 0
    
    
    for i in range(n):
        total += x[i]*y[i]  # Multiple each iration of the 2 input vectors and total the sum.
        
    for i in range(n):
        x_squared += x[i]**2   # Sum the square of each iteration of the input vector x.
    
    ans = (total-(n*x_bar*y_bar))/(x_squared-(n*x_squared_bar))   
    
    return ans
```

```python
def B0(x,y):   # Function to compute B0 coefficient.
    
    x_bar = np.mean(x)
    y_bar = np.mean(y)
    
    ans = y_bar - B1(x,y)*x_bar
    
    return ans
```

```python
# 2 Vector to test the vector with.
x = [1,4,6,5,4,7,8,9,0,5,4,3,2,4,5,-4,-9]
y = [2,4,5,5,4,8,10,8,0,5,3,4,2,5,6,-5,-8]
```

```python
plt.scatter(x,y)  # Scatter plot fo the 2 variables.
```

```python
def Linear_regression_model(x,y):    # Function to plot the least square regression line.
    
    x_estimate = np.linspace(min(x),max(x),100)
    
    y_estimate = B0(x,y) + B1(x,y)*x_estimate
    print(B0(x,y))
    
    return plt.plot(x_estimate,y_estimate), plt.scatter(x,y) , plt.scatter(np.mean(x),np.mean(y))
```

```python
Linear_regression_model(x,y)
```

```python
def predict(x,y,x_test):   # Function to predict a value using the model.
    
    y_predict = B0(x,y) + B1(x,y)*x_test
    
    return y_predict
```

```python
Linear_regression_model(x,y)   # Plot the linear regression model
plt.scatter(-3,predict(x,y,-3),color='red')   # Plot the predicted value in red. We can see the model is accurate.
```
